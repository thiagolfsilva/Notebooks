```python
import requests
import json
```

```python
BASE_KUCOIN='https://api.kucoin.com'
```

```python
#Gets all the currencies on Kucoin that have margin trading availabe
#response = requests.get(BASE_KUCOIN+'/api/v1/currencies')
response = requests.get('https://api.kucoin.com/api/v1/currencies')
data=response.json()
currencies = []
for datapoint in data['data']:
    if datapoint['isMarginEnabled']:
        currencies.append(datapoint['currency'])

currencies.sort()
print(currencies)        
```

```python
def marginal_rates(coinset, USD_threshold, showprogress=False):
    aprs=[]
    for coin in coinset:
        if showprogress:
            print(coin)
        response = requests.get(BASE_KUCOIN+'/api/v1/prices?currencies='+coin)
        data=response.json()
        price=float(data['data'][coin])
        response = requests.get(BASE_KUCOIN+'/api/v1/margin/market?currency='+coin)
        data=response.json()

        borrowable=False
        total=0
        
        for datapoint in data['data']:
            total+=float(datapoint['size'])
            if total*price > USD_threshold:
                APR = float(datapoint['dailyIntRate'])*365
                borrowable=True
                break
        if borrowable:
            aprs.append([coin, APR])
        else:
            aprs.append([coin, -1])
    return aprs
```

```python
aprs=marginal_rates(currencies, 2000)
```

```python
for apr in aprs:
    print(apr[0]+': {:.2%}'.format(apr[1]))
```

```python
#Sort by alphabetical order
#okx_aprs.sort(key=lambda x:x[0])

#Sort by rate
aprs.sort(key=lambda x:x[1])

for apr in aprs:
    print(apr[0]+': {:.2%}'.format(apr[1]))
```

```python
SET1_KCS=['ALICE',
          'AXS',
          'CAKE',
         'CLV',
         'CRO',
         'CRV',
         'FLOW',
          'GLMR',
          'KAVA',
          'MOVR',
         'ROSE',
          'SLP',
          'STX',
         'VET',]

rates=marginal_rates(SET1_KCS, 2000)
for apr in rates:
    print(apr[0]+': {:.2%}'.format(apr[1]))
```
